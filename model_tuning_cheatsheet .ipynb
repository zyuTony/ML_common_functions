```python
# logistic Regression
param_grid = {'penalty' : ['l1', 'l2'],
              'C' : np.logspace(-4, 4, 20),
              'solver' : ['liblinear']}

# r2 and accuracy for default scoring
logreg_grid = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring="f1", verbose=2) 
logreg_grid.fit(X_train, y_train)

logreg_grid.predict(X_test,y_test)
print(logreg_grid.best_params_, logreg_grid.best_score_)
```

```python
# SVC, SVR --- kernel=poly + gamma=1 + C=1  takes a long time to train
param_grid = {'C': [0.1, 1, 10, 100], 
              'gamma': [1,0.1,0.01,0.001],
              'kernel': ['rbf', 'poly', 'sigmoid']}

svm_grid = GridSearchCV(SVC(), param_grid, scoring='f1', refit=True, cv=5, verbose=2)
svm_grid.fit(X_train,y_train)

svm_grid.predict(X_test,y_test)
print(svm_grid.best_params_, svm_grid.best_score_)
```

```python
# Random Forest 
param_grid = {'n_estimators': [200, 300, 400], # num of trees
              'max_features': ['auto', 'sqrt'], # features used at each node
              'bootstrap': [True, False], # bootstrap data or not
              'min_samples_split': [2, 5, 10], # num of samples to split
              'min_samples_leaf': [1, 2, 4], # min sample in each leaf node   
             }

rf_grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, 
                       verbose=2)
rf_grid.fit(X_train,y_train)

# Random CV
rf_grid = RandomizedSearchCV(RandomForestRegressor(), param_grid, cv=5, 
                             verbose=2, n_iter = 100, n_jobs=-1)
rf_grid.fit(X_train,y_train)

rf_grid.predict(X_test,y_test)
print(svm_grid.best_params_, svm_grid.best_score_)
```
